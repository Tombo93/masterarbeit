```python
import os

import torch
import numpy as np
from torchvision import transforms

from data.dataset import Cifar10BackdoorDataset
```

```python
cifar10_data_path = os.path.abspath(os.path.join(os.path.dirname("predict_poison_labels.ipynb"), os.pardir, "data"))
test_data_path = os.path.join(cifar10_data_path, "processed", "cifar10", "backdoor-cifar10-test.npz")
test_transform = transforms.Compose(
    [transforms.ToTensor(), transforms.Normalize([0.4914, 0.4822, 0.4465], [0.2023, 0.1994, 0.2010])]
        )
backdoor_dataset = Cifar10BackdoorDataset(test_data_path, test_transform)
testloader = torch.utils.data.DataLoader(
        backdoor_dataset,
        batch_size=64,
        shuffle=False,
        num_workers=2,
        pin_memory=True,
    )
```

```python
def is_poison_label(pred, poison_class=9):
    if pred == poison_class:
        return 1
    return 0
```

```python
labels = backdoor_dataset.labels
predicted_poison = map(is_poison_label, labels)
np.bincount(list(predicted_poison))
```
