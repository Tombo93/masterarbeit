```python
import os
import copy

import pandas as pd
import matplotlib.pyplot as plt
```

```python
report_path = os.path.abspath(os.path.join(os.path.dirname("cifar10.ipynb"), os.pardir, "reports", "cifar10"))
clean_report = os.path.join(report_path, "by-class-clean-data-test.csv")
backdoor_report = os.path.join(report_path, "by-class-backdoor-test.csv")


clean_df = pd.read_csv(clean_report)
backdoor_df = pd.read_csv(backdoor_report)
```

```python
backdoor_df.head()
```

```python
ax1 = backdoor_df.plot(x='Unnamed: 0', y=['plane','car','bird','deer','dog','frog','horse','ship','truck'])
ax1.set_xlabel("epochs")
ax1.set_ylabel("accuracy")
ax1.set_title("backdoored data")
ax2 = clean_df.plot(x='Unnamed: 0', y=['plane','car','bird','deer','dog','frog','horse','ship','truck'])
ax2.set_xlabel("epochs")
ax2.set_ylabel("accuracy")
ax2.set_title("clean data")
```

```python
detect_backdoor_report = os.path.join(report_path, "backdoor-test.csv")
detect_backdoor_df = pd.read_csv(detect_backdoor_report)
ax = detect_backdoor_df.plot(x='Unnamed: 0', y=['BinaryAccuracy','BinaryRecall','BinaryPrecision'])
ax.set_xlabel("epochs")
ax.set_title("backdoored label detection")
ax.set_ylim(bottom=0.0, top=1.0)
```
